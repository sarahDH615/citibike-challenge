```python
# 2021 csvs
jan2021 = pd.read_csv('bike_csvs/JC-202101-citibike-tripdata.csv')
feb2021 = pd.read_csv('bike_csvs/JC-202102-citibike-tripdata.csv')
mar2021 = pd.read_csv('bike_csvs/JC-202103-citibike-tripdata.csv')
```

```python
data2021 = pd.concat([jan2021, feb2021, mar2021]).reset_index(drop=True)
print(data2021.shape)
print(data2021.columns)
data2021.sample(5)
```

```python
data18to21 = pd.concat([data2018, data2019, data2020, data2021]).reset_index(drop=True)
data18to21.sample(5)
```

```python
# investigating whether there are suspicious data points
data18to21['birth year'].unique()
```

```python
# filtering out dobs that would make the user greater than 90 years old
d1821 = data18to21[data18to21['birth year'] >= 1931]
print(d1821['birth year'].unique())
d1821.sample(5)
```

```python
# copying index to a column to create a unique id for each trip
d1821['tripID'] = d1821.index
d1821 = d1821[['tripID', 'tripduration', 'starttime', 'stoptime', 'start station id',
    'start station name', 'start station latitude',
    'start station longitude', 'end station id', 'end station name',
    'end station latitude', 'end station longitude', 'bikeid', 'usertype',
    'birth year', 'gender']]
d1821.sample(5)
```

```python
# exporting full dataset
d1821.to_csv('bike_csvs/data1821NEW.csv', index=False)
```
